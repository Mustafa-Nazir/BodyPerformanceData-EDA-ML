## Arrange the columns for prediction

```python
rowData["class"].unique()
```

```python
data = rowData.copy()
data.loc[data["class"] == "D","class"] = 1
data.loc[data["class"] == "C","class"] = 2
data.loc[data["class"] == "B","class"] = 3
data.loc[data["class"] == "A","class"] = 4
data["class"] = data["class"].astype("int")
data.info()
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

## Feature Engineering
## Calculating and Adding BMI

```python
indexData = data.copy()
indexData["BMIndex"] = indexData["weight_kg"] / (indexData["height_cm"]/100)**2
indexData["situation"] = None
indexData.loc[indexData["BMIndex"] < 18.5 , "situation"] = 1 #underweight
indexData.loc[(indexData["BMIndex"] >= 18.5)&(indexData["BMIndex"] < 25) , "situation"] = 2 #normal
indexData.loc[(indexData["BMIndex"] >= 25)&(indexData["BMIndex"] < 30) , "situation"] = 3 #owerweight
indexData.loc[(indexData["BMIndex"] >= 30)&(indexData["BMIndex"] < 35) , "situation"] = 4 #obese
indexData.loc[indexData["BMIndex"] >=35 , "situation"] = 5 #extremely-obese
indexData["situation"] = indexData["situation"].astype("int")
indexData.head(15)
```

```python
indexData.info()
```

## Calculating and Adding Diastolic and Systolic

```python
#For Systolic
indexData["_systolic"] = None
indexData.loc[(indexData["systolic"] >= 120) & (indexData["systolic"] <= 130) , "_systolic"] = 2 #normal
indexData.loc[(indexData["systolic"] < 120) , "_systolic"] = 1 #low
indexData.loc[(indexData["systolic"] > 130) , "_systolic"] = 3 #high
indexData["_systolic"] = indexData["_systolic"].astype("int")
```

```python
#For Diastolic
indexData["_diastolic"] = None
indexData.loc[(indexData["diastolic"] >= 70) & (indexData["diastolic"] <= 90) , "_diastolic"] = 2 #normal
indexData.loc[(indexData["diastolic"] < 70) , "_diastolic"] = 1 #low
indexData.loc[(indexData["diastolic"] > 90) , "_diastolic"] = 3 #high
indexData["_diastolic"] = indexData["_diastolic"].astype("int")
```
